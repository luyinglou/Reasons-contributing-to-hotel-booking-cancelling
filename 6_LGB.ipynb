```
import numpy as np
from sklearn.metrics import accuracy_score,recall_score,confusion_matrix

xgbc = lgb.LGBMRegressor()

param_list = {
        'n_estimators':range(100,1000,100),
        'learning_rate':np.linspace(0.01,0.2,30),
        'min_data_in_leaf':range(5,50,2),
        'max_depth':range(2,30,2),
        'reg_alpha':np.linspace(0,3,10)
        }

lgb_grid = RandomizedSearchCV(xgbc,param_list,cv = 3,scoring = 'neg_mean_squared_error',n_iter=300,n_jobs = -1)

lgb_grid.fit(X_train.values,np.ravel(y_train))
lgbr_best_estimator = lgb_grid.best_estimator_
```

```
preds = lgbr_best_estimator.predict(X_test.values)
predictions_class =  np.where(preds>0.5, 1, 0)
acc_lgb = accuracy_score(y_test, predictions_class)
print('Accuracy with best LGB parameters is:', acc_lgb)
```

```
# selecting cutoffs
import matplotlib.pyplot as plt
import seaborn as sns
cutoffs = np.linspace(0.3, 1.0, 70)
acc = []
for c in cutoffs:
  predictions_class =  np.where(preds > c, 1, 0)
  acc_lgb = accuracy_score(y_test, predictions_class)
  acc.append(acc_lgb)
plt.plot(cutoffs, acc)
plt.xlabel('cutoffs')
plt.ylabel('Test Accuracy')
display()
```

####Draw auc graph

```
from matplotlib import pylab
# Draw R/P Curve
def plot_pr(auc_score, precision, recall, label=None):
    pylab.figure(num=None, figsize=(6, 5))
    pylab.xlim([0.0, 1.0])
    pylab.ylim([0.0, 1.0])
    pylab.xlabel('Recall')
    pylab.ylabel('Precision')
    pylab.title('P/R (AUC=%0.2f) / %s' % (auc_score, label))
    pylab.fill_between(recall, precision, alpha=0.5)
    pylab.grid(True, linestyle='-', color='0.75')
    pylab.plot(recall, precision, lw=1)
    pylab.show()
```
