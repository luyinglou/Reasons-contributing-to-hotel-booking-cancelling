#### Market segment

```
##calculate the market segments for each hotel
import numpy as np
res_hotel =  hotel_DF[ hotel_DF['hotel']=='Resort Hotel']
m = pd.DataFrame(res_hotel['market_segment'].value_counts())
l = m.loc[['Online TA','Offline TA/TO','Groups','Direct','Corporate','Complementary']]
x1=np.array(l.index)
y1=np.array(l.market_segment)

city_hotel =  hotel_DF[ hotel_DF['hotel']=='City Hotel']
n = pd.DataFrame(city_hotel['market_segment'].value_counts())
x2=np.array(n.index)
y2=np.array(n.market_segment)
```

```
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
sns.set_style("white")
plt.pie(y1,labels=x1, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
display()
```

####RH: Market in different countries

```
res_hotel =  hotel_DF[ hotel_DF['hotel']=='Resort Hotel']
t = pd.DataFrame(res_hotel['country'].value_counts())
s = t.head(10)
s.insert(1,'percent',s['country']/(t['country'].sum()))
```

```
s.percent
```

```
percent = s.percent.round(decimals=2)
```

```
#Resort Hotel's market in different countries
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns

sns.set(rc={'figure.figsize':(8,4)})
axis = sns.barplot(x=s.index, y=percent)
axis.set_title("Resort Hotel's Top 10 markets",fontsize= 15)
axis.set_ylabel('Percent of total bookings',fontsize=10)
axis.set_xlabel('Country',fontsize=10)

for p in axis.patches:
    height = p.get_height()
    axis.annotate('{}'.format(height),xy=(p.get_x() + p.get_width() / 2, height), ha='center', va='bottom',fontsize=10)
display(axis)
```
