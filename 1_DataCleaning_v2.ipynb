```
import pandas as pd
import numpy as np
```

####Import Dataset and Drop unneccesary columns

```
delimiter=','
hotelDF = (spark.read
  .option('sep', delimiter)
  .option("header", True)
  .option("inferSchema", True)
  .csv("/mnt/s3data/hotel_bookings.csv")
  .drop('reservation_status','distribution_channel','reservation_status_date')
)
display(hotelDF)
```

```
display(hotelDF.describe())
```

####Data Cleaning

```
# filter out unreasonable data where 'adr' < 0
from pyspark.sql.functions import col
hotelDF=hotelDF.filter(col("adr")>=0)
```

```
# convert 'children' to float type
from pyspark.sql.types import DoubleType
hotelDF = hotelDF.withColumn('children_num',hotelDF['children'].cast(DoubleType())).drop('children')
```

```
# use pandas for furthur cleaning
df = hotelDF.toPandas()
```

```
# convert strings like "NULL" and "NA" to real null values
cols_with_null = ['children_num', 'country', 'agent', 'company']
for i in cols_with_null:
  df[i] = df[i].apply(lambda x: np.nan if x in ['NULL', 'NA'] else x)
```

```
# impute mean if 'children' is null
children_mean = np.floor(df[df['children_num'] > 0]['children_num'].mean())
df['children_num'] = df['children_num'].apply(lambda x: children_mean if pd.isnull(x) else x)
```

```
# impute mean if 'adult' is 0'
adult_mean = np.floor(df[df['adults'] > 0]['adults'].mean())
df['adults'] = df['adults'].apply(lambda x: adult_mean if x == 0 else x)
```

```
# impute 'adr' of 0 as the mean of different customer type
df['adr'] = df['adr'].apply(lambda x: None if x <= 0 else x)
df['adr'] = df['adr'].fillna(df.groupby('customer_type')['adr'].transform('mean'))
```

```
# impute 'unknown' if 'country' is null
df['country'] = df['country'].apply(lambda x: 'unknown' if pd.isnull(x) else x)
```

```
# get rid of outliers
df = df[df['adr'] <= 500]
```

```
# create new variable 'if_agent', representing if the booking was done by an agent
df['if_agent'] = df['agent'].apply(lambda x: 0 if pd.isnull(x) else 1)
```

```
# create new variable 'if_company'. representing if the booking was done by a company
df['if_company'] = df['company'].apply(lambda x: 0 if pd.isnull(x) else 1)
```

```
# define a function to decide if 'reserved_room_type' is the same as 'assigned_room_type'
def room_type(cols):
    reserved = cols[0]
    assigned = cols[1]
    if reserved == assigned:
        return 1
    else:
        return 0
    

df['if_desired_room'] = df[['reserved_room_type', 'assigned_room_type']].apply(room_type, axis=1)
```

```
hotelDF1 = spark.createDataFrame(df) 
display(hotelDF1)
```
