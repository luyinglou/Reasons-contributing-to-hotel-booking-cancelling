#### Cancellation Rate in Different Countries

```
from pyspark.sql.functions import col
res_hotelDF = hotelDF.filter(col("hotel")!='Resort Hotel')
city_hotelDF = hotelDF.filter(col('hotel')!='City Hotel')
```

```
booking_country = (res_hotelDF
  .select('country','is_canceled')
  .groupBy("country")
  .count()
  .orderBy('count',ascending=False)
)
city_booking_country = (city_hotelDF
  .select('country','is_canceled')
  .groupBy("country")
  .count()
  .orderBy('count',ascending=False)
)
```

```
res_bookingcancel_country = (res_hotelDF
  .select('country','is_canceled')
  .groupBy("country")
  .sum()
)

res_cancel_country = res_bookingcancel_country.join(booking_country,on=['country'], how='inner')
```

```
from pyspark.sql.functions import col,round
res_cancelrate_country = res_cancel_country.select('country',round(col('sum(is_canceled)')/col('count'),2))
display(res_cancelrate_country)
```

```
city_bookingcancel_country = (city_hotelDF
  .select('country','is_canceled')
  .groupBy("country")
  .sum()
)

city_cancel_country = city_bookingcancel_country.join(city_booking_country,on=['country'], how='inner')

from pyspark.sql.functions import col,round
city_cancelrate_country = city_cancel_country.select('country',round(col('sum(is_canceled)')/col('count'),2))
display(city_cancelrate_country)
```

#### Cancel rate and total booking number over the months

```
import numpy as np
import pandas as pd
sns.set(rc={'figure.figsize':(8,4)})
sns.set_style("white")
hotel = hotelDF.toPandas()
book_bymonth = pd.DataFrame(hotel['hotel'].groupby(by=hotel['arrival_date_month']).count())
cancel_bymonth = pd.DataFrame(hotel['is_canceled'].groupby(by=hotel['arrival_date_month']).sum())
cancel_rate = pd.merge(cancel_bymonth, book_bymonth, left_index=True, right_index=True, how='left')
cancel_rate.insert(2,'cancel rate',cancel_rate['is_canceled']/cancel_rate['hotel'])
cancel_rate = cancel_rate.loc[['January','February','March','April','May','June','July','August','September','October','November','December']]

fig, ax1 = plt.subplots(figsize=(8, 4))
ax2 = ax1.twinx()
cancel_rate['hotel'].plot(kind='bar', color='m', ax=ax1)
cancel_rate['cancel rate'].plot(kind='line',color='blue', marker='^',ax=ax2)
ax1.yaxis.tick_right()
ax2.yaxis.tick_left()
ax1.set_title('Cancellation rate over the months')
display()
```
