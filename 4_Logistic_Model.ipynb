#### Logistic regression: L1, L2

```
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn import  metrics as mt
from sklearn.linear_model import LogisticRegression, Ridge, Lasso
import numpy as np

X_train, X_test, y_train, y_test = (train_test_split(hotelDF.drop(["is_canceled"], axis=1), hotelDF[["is_canceled"]].values.ravel(),test_size=0.3, random_state=42))
```

```
lgr_model1 = LogisticRegression(penalty='l1',solver='liblinear',verbose=0) 
lgr_model1.fit(X_train,y_train)
predictY1 = lgr_model1.predict(X_test) 
acc1 = mt.scorer.accuracy_score(y_test,predictY1)
acc1
```

```
lgr_model2 = LogisticRegression(penalty='l2',verbose=0) 
lgr_model2.fit(X_train,y_train)
predictY2 = lgr_model2.predict(X_test) 
acc2 = mt.scorer.accuracy_score(y_test,predictY2)
acc2
```

####Grid Search

```
from sklearn.metrics import accuracy_score,recall_score,confusion_matrix

parameters = {'C': [1e-10, 1e-8, 1e-4, 1e-3,1e-2, 0.1, 1, 5, 10, 20, 30]}

lgr_model_1 = LogisticRegression(penalty='l1',solver='liblinear',verbose=0)

gridlasso = GridSearchCV(lgr_model_1, parameters, cv=10)

#fit the models
gridlasso.fit(X_train, y_train)

print("lasso best parameters:", gridlasso.best_params_)
print("lasso score:", gridlasso.score(X_test, y_test))
print("lasso best estimator coef:", gridlasso.best_estimator_.coef_)
test_pred = gridlasso.best_estimator_.predict(X_test)
print(confusion_matrix(y_test,test_pred))
```

```
# selecting cutoffs
test_preds_prob1 = gridlasso.best_estimator_.predict_proba(X_test)
test_preds_prob1 = np.array(list(map(lambda x:x[1], test_preds_prob1)))

import matplotlib.pyplot as plt
import seaborn as sns
cutoffs = np.linspace(0, 1.0, 100)
acc = []
for c in cutoffs:
  predictions_class =  np.where(test_preds_prob1 > c, 1, 0)
  acc_lasso = accuracy_score(y_test, predictions_class)
  acc.append(acc_lasso)
  

test_preds_prob
max_index = acc.index(max(acc))
best_cutoff = cutoffs[max_index]
print("The best cutoff: ", best_cutoff)
print("lasso model with the best coefficient for has the accuracy of {}".format(max(acc)))
```

```
sns.set(rc={'figure.figsize':(5,3)})
plt.plot(cutoffs, acc)
plt.xlabel('cutoffs')
plt.ylabel('Test Accuracy')
display()
```

```
parameters = {'C': [  1e-4, 1e-3,1e-2, 0.1, 1, 5, 10, 20, 30, 40]}

lgr_model_2 = LogisticRegression(penalty='l2',solver='liblinear',verbose=0) 

gridridge = GridSearchCV(lgr_model_2, parameters,  cv=10)

#fit the models
gridridge.fit(X_train, y_train)

print("ridge best parameters:", gridridge.best_params_)
print("ridge score:", gridridge.score(X_test, y_test))
print("ridge best estimator coef:", gridridge.best_estimator_.coef_)
test_pred = gridridge.best_estimator_.predict(X_test)
print(confusion_matrix(y_test,test_pred))
```

#### Coefficient of the best logistic regression model

```
cols = hotelDF.columns
col_dict={}
for i in range(0,28):  ##interpret the coefficients
    col_dict[gridlasso.best_estimator_.coef_[0][i]]=cols[i+1]
    print("β{} (coefficient for {} ): {}".format((i+1),cols[i+1],gridlasso.best_estimator_.coef_[0][i]))
```

```
c = sorted(col_dict.keys())
c_negative = c[0:8]
c_positive = c[-5:]
c_negative_cols = [col_dict[i] for i in c_negative]
c_positive_cols = [col_dict[i] for i in c_positive]
for i in range(0,8):
  print("coefficient for {} : {}".format(c_negative_cols[i], c_negative[i]))
  
for i in range(0,5):
  print("coefficient for {} : {}".format(c_positive_cols[i], c_positive[i]))
```

####Important Factors

```
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns

sns.set(rc={'figure.figsize':(25,6)})
axis = sns.barplot(x=c_negative_cols, y=[-i for i in c_negative])
axis.set_title("The important negative factors",fontsize= 20)
axis.set_ylabel('Coefficients',fontsize=15)
axis.set_xlabel('Factors',fontsize=15)
for p in axis.patches:
    height = p.get_height()
    axis.annotate('{}'.format(height),xy=(p.get_x() + p.get_width() / 2, height), ha='center', va='bottom',fontsize=10)

display(axis)
```

```
sns.set(rc={'figure.figsize':(25,6)})
axis = sns.barplot(x=c_positive_cols, y=c_positive)
axis.set_title("The important positive factors",fontsize= 20)
axis.set_ylabel('Coefficients',fontsize=15)
axis.set_xlabel('Factors',fontsize=15)
for p in axis.patches:
    height = p.get_height()
    axis.annotate('{}'.format(height),xy=(p.get_x() + p.get_width() / 2, height), ha='center', va='bottom',fontsize=10)

display(axis)
```
