```
import pandas as pd

hotelDF = spark.read.parquet("dbfs:/mnt/s3data/hotel_booking_Indexcleaned.parquet/").toPandas()
display(hotelDF)
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(hotelDF.drop(["is_canceled"], axis=1), hotelDF[["is_canceled"]].values.ravel(), test_size = 0.25, random_state=42)
```

#### Train the model

```
import mlflow.sklearn
from sklearn.ensemble import RandomForestClassifier
rf_hotel = RandomForestClassifier(n_estimators=100, criterion='gini', max_depth=None, bootstrap=True) 
```

```
from sklearn.metrics import accuracy_score,recall_score,confusion_matrix
rf_hotel.fit(X_train, y_train)
test_pred = rf_hotel.predict(X_test)
print(confusion_matrix(y_test,test_pred))
print(accuracy_score(y_test,test_pred))
```

#### Cross validation by estimators

```
from sklearn.model_selection import KFold, cross_val_score
n_folds = 5
def acc_cv(model):
    kf = KFold(n_folds, shuffle=True,
               random_state=42).get_n_splits(X_train)
    acc = cross_val_score(model, X_train, y_train,scoring="accuracy", cv=kf)
    return (acc)
```

```
import mlflow.sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, make_scorer
import numpy as np
# dictionary containing hyperparameter names and list of values we want to try
parameters = {'n_estimators': [10, 100, 200, 500, 1000] , 
              'max_depth': [None, 10,50,100,200,500], 
              'bootstrap': [False, True]}

rf = RandomForestClassifier()
grid_rf_model = GridSearchCV(rf, parameters,scoring='accuracy', cv=5, n_jobs=-1)
grid_rf_model.fit(X_train, y_train)

best_rf = grid_rf_model.best_estimator_
for p in parameters:
  print("Best '{}': {}".format(p, best_rf.get_params()[p]))
```

```
from sklearn.metrics import accuracy_score,recall_score,confusion_matrix
test_pred = best_rf.predict(X_test)
print(confusion_matrix(y_test,test_pred))
print(accuracy_score(y_test,test_pred))
```

#### Feature Importance

```
names=hotelDF.columns[1:]
Feature = sorted(zip(map(lambda x: round(x, 4), best_rf.feature_importances_), names), reverse=True)
for i in Feature:
    print(i)
Best_RFdict = dict(Feature)
```

```
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns

sns.set(rc={'figure.figsize':(25,6)})
axis = sns.barplot(x=list(Best_RFdict.values()), y=list(Best_RFdict.keys()))
axis.set_title("The Importance of Feature",fontsize= 20)
axis.set_ylabel('Importance',fontsize=15)
plt.xticks(rotation='vertical');

for p in axis.patches:
    height = p.get_height()
    axis.annotate('{}'.format(height),xy=(p.get_x() + p.get_width() / 2, height), ha='center', va='bottom',fontsize=10)

display(axis)
```
